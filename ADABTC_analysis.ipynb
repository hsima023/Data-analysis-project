# Setting up Model
First, we use 'volume' to predict the 'open' price.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = df['volume'].to_numpy()
Y = df['open'].to_numpy()

X=np.reshape(X,(-1,1))
Y=np.reshape(Y,(-1,1))

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state= 42) 
```

```python
# Train linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Train 100-nearest neighbors model
knn_model = make_pipeline(
    StandardScaler(),
    KNeighborsRegressor(n_neighbors=100)
)
knn_model.fit(X=X_train, y=y_train)
```

```python
ensemble_model = VotingRegressor([
    ("linear", linear_model), 
    ("knn", knn_model)
])
ensemble_model.fit(X=X_train, y=y_train)
```

```python
from sklearn.model_selection import cross_val_score

for model in [linear_model, knn_model,ensemble_model]: # models 
    print(-cross_val_score(model, X=X_train, y=y_train, cv=5,
                           scoring="neg_mean_squared_error").mean())
```

As we can see, the ensemble model has the lowest value. Hence, the ensemble model is the best model among the three models. Let's do a model evaluation using the ensemble model and calculate the R-squared error from the predicted value.

```python
prediction = ensemble_model.predict(X_test)

error_value = metrics.r2_score(y_test, prediction)
print('R squared error:', error_value)
```

It seems like using only volume is a bad idea to predict the price. Let's try other features 'trades'.

```python
X = df['trades'].to_numpy()
Y = df['open'].to_numpy()

X=np.reshape(X,(-1,1))
Y=np.reshape(Y,(-1,1))

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state= 42) 

# Train linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Train 10-nearest neighbors model
knn_model = make_pipeline(
    StandardScaler(),
    KNeighborsRegressor(n_neighbors=100)
)
knn_model.fit(X=X_train, y=y_train)

# voting ensemble model
ensemble_model = VotingRegressor([
    ("linear", linear_model), 
    ("knn", knn_model)
])
ensemble_model.fit(X=X_train, y=y_train)

for model in [linear_model, knn_model,ensemble_model]: # models 
    print(-cross_val_score(model, X=X_train, y=y_train, cv=5,
                           scoring="neg_mean_squared_error").mean())
```

The result indicates that using knn-model is the best choice. The MSE value is even lower that the MSE value for feature 'volume'. Let's do a model evaluation using the knn model and calculate the R-squared error from the predicted value.

```python
prediction = ensemble_model.predict(X_test)

error_value = metrics.r2_score(y_test, prediction)
print('R squared error:', error_value)
```

This score is better than the score of feature 'volume'. We can conclude that using 'trades' is better than using 'volume' to predict the price. Now, let's setup a model using multiple features.
